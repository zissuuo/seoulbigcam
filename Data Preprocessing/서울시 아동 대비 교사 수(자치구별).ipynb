## 아동 대비 교사 수

```python
import pandas as pd
import numpy as np
```

```python
# 데이터 로드

kin = pd.read_csv("[교육]서울시 유치원 통계(2021-2022).csv")
kin.head()
kin = kin.drop(columns=['2021_kin', '2022_kin'])
```

```python
# 비율 컬럼 생성

kin['2021_tea_rate'] = kin['2021_st'] / kin['2021_tea']
kin['2022_tea_rate'] = kin['2022_st'] / kin['2022_tea']
kin = kin.drop(columns = ['2021_tea', '2021_st', '2022_tea', '2022_st'])
kin.head()
```

```python
# 2021 교사 대비 아동 수

kin.sort_values(by = '2021_tea_rate')
```

```python
# 2022 교사 대비 아동 수

kin.sort_values(by = '2022_tea_rate')
```

```python
# 변화량

kin['variation'] = kin['2022_tea_rate'] - kin['2021_tea_rate']
kin.sort_values(by = 'variation', ascending = False)
```
