## 보육시설

### 동별

```python
# CSV 파일 로드
import pandas as pd

fac = pd.read_csv('../dataset/edu/서울시 행정동별 보육시설 형태(2021-2022).csv')
fac.head()
```

```python
# 동별 비율 컬럼 생성
fac['2022_pub_per'] = fac['2022_pub'] / fac['2022_all'] * 100
fac_new = fac[['구', '동별', '2022_pub_per']]
fac_new.head()
```

```python
# 2022 국공립보육시설 비율 하위 10개
fac_new.sort_values(by = '2022_pub_per').head(15)
```

```python
fac_new.to_csv('../dataset/edu/서울시 행정동별 국공립어린이집 비율.csv')
```

### 구별

```python
# 구별 데이터 확인
fac_gu = fac.groupby('구', as_index=False).sum()

# 비율 컬럼 드랍
fac_gu = fac_gu.drop(columns = ['2022_pub_per'])
fac_gu.head()
```

```python
# 구별 비율 데이터 확인
fac_gu['2022_pub_per'] = fac_gu['2022_pub'] / fac_gu['2022_all'] * 100
fac_gu = fac_gu.drop(columns = ['2021_all', '2021_pub', '2022_all', '2022_pub'])

fac_gu.sort_values(by = '2022_pub_per')
```

```python
fac_gu.to_csv('../dataset/edu/서울시 자치구별 국공립어린이집 비율.csv')
```
