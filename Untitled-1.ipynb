```python
import pandas as pd
```

```python
cluster = pd.read_csv('cluster1.csv')
```

```python
chi = pd.read_csv('/Users/kimjisu/Desktop/seoulbigcam/Final Data/교육_국공립어린이집 비율_동별_2022.csv')
kin = pd.read_csv('/Users/kimjisu/Desktop/seoulbigcam/Final Data/교육_국공립유치원 비율_동별_2023.csv')
```

```python
df = pd.merge(chi, kin, on = ['구', '동'], how = 'outer')
```

```python
df.columns
```

```python
df.rename(columns = {'전체': 'all_kin', '국공립': 'pub_kin', '국공립 어린이집 비율': 'ratio_kin',
'합계': 'all_chi', '공립 수': 'pub_chi', '국공립 유치원 비율': 'ratio_chi'}, inplace = True)
df = df.fillna(0)
df
```

```python
df.to_csv('유치원보육시설.csv')
```

```python
gu = []
dong = []

data = cluster['행정구역'].tolist()

for item in data:
    parts = item.split('_')
    gu.append(parts[0])
    dong.append(parts[1])
```

```python
df_cluster = {'구': gu, '동': dong}
df_cluster = pd.DataFrame(df_cluster)
```

```python
df_cluster
```

```python
cl1 = pd.merge(df_cluster, df, on = ['구', '동'], how = 'left')
cl1
```
