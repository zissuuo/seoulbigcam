```python
import pandas as pd

def min_max_scaling(dataframe, feature_range=(0, 1)):
    """
    Perform Min-Max scaling on each column of a DataFrame.

    Parameters:
    - dataframe: The input DataFrame.
    - feature_range: A tuple specifying the desired scaling range (default is 0 to 1).

    Returns:
    - scaled_dataframe: A new DataFrame with scaled values.
    """
    # Copy the input DataFrame to avoid modifying the original data
    scaled_dataframe = dataframe.copy()

    # Calculate the minimum and maximum values for each column
    min_vals = dataframe.min()
    max_vals = dataframe.max()

    # Calculate the feature range
    min_val, max_val = feature_range

    # Perform scaling for each column
    for column in dataframe.columns:
        scaled_dataframe[column] = (dataframe[column] - min_vals[column]) / (max_vals[column] - min_vals[column]) * (max_val - min_val) + min_val

    return scaled_dataframe
```

```python
house2022 = pd.read_csv('2022서울시 부동산 실거래가 정보.csv', encoding='cp949')
house2021 = pd.read_csv('2021서울시 부동산 실거래가 정보.csv', encoding='cp949')
house2020 = pd.read_csv('2020서울시 부동산 실거래가 정보.csv', encoding='cp949')
```

```python
house = pd.concat([house2022, house2021, house2020])
```

```python
house.info()
```

```python
house = house[['자치구명', '법정동명', '물건금액(만원)', '건물면적(㎡)']]
```

```python
house['mean_vlue/m**2'] = house['물건금액(만원)'] / house['건물면적(㎡)']
```

```python
house = house.groupby('법정동명')['mean_vlue/m**2'].mean()
```

```python
house.to_csv('house_values.csv', encoding='utf-8')
```

```python
def convert_exponential_to_float(exponential_str):
    try:
        return float(exponential_str)
    except ValueError:
        return exponential_str

# 데이터프레임의 특정 열에 함수를 적용
money['공시지가(원/㎡)'] = money['공시지가(원/㎡)'].apply(convert_exponential_to_float)
```
